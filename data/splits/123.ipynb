```python
import pandas as pd

train = pd.read_csv("train.csv")
val   = pd.read_csv("val.csv")
test  = pd.read_csv("test.csv")

OOS_CODE = 42  # код oos в clinc_oos

print("=" * 70)
print("РАЗМЕРЫ ВЫБОРОК (для таблицы 1)")
print("=" * 70)
for name, df in [("Обучающая", train), ("Валидационная", val), ("Тестовая", test)]:
    total   = len(df)
    oos     = (df["intent"] == OOS_CODE).sum()
    domain  = total - oos
    n_classes_dom = df[df["intent"] != OOS_CODE]["intent"].nunique()
    per_class = round(domain / n_classes_dom, 1)
    print(f"{name:15s}: всего={total:5d}, доменных={domain:5d}, "
          f"OOS={oos:4d}, ~примеров на доменный класс={per_class}")

print()
print("=" * 70)
print("РАСПРЕДЕЛЕНИЕ ПО ДОМЕНАМ В ОБУЧАЮЩЕЙ ВЫБОРКЕ (для подраздела 2.2.1)")
print("=" * 70)
# Маппинг intent -> domain
# CLINC150 plus имеет фиксированную таксономию.
# Простой способ: считать число примеров на каждый класс,
# а домены вы знаете отдельно.
# Здесь только число примеров на каждый интент в train:
counts = train[train["intent"] != OOS_CODE]["intent"].value_counts()
print(f"Примеров на доменный класс: min={counts.min()}, "
      f"max={counts.max()}, median={counts.median():.1f}")
print(f"Уникальных доменных классов в train: {counts.shape[0]}")
print(f"OOS-примеров в train: {(train['intent']==OOS_CODE).sum()}")

print()
print("=" * 70)
print("ДЛИНЫ ЗАПРОСОВ (для подраздела 2.2.2)")
print("=" * 70)
all_text = pd.concat([train["text"], val["text"], test["text"]])
lens = all_text.str.split().str.len()
print(f"min={lens.min()}, max={lens.max()}, "
      f"медиана={lens.median():.1f}, среднее={lens.mean():.2f}")
print(f"квантили: 25%={lens.quantile(0.25):.0f}, "
      f"75%={lens.quantile(0.75):.0f}, 95%={lens.quantile(0.95):.0f}")
```
